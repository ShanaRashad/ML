## Titanic Survival

```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/user/Documents/Machine Learning/unsupervised/PROJECT/Titanic.csv")
df.head()
```

```python
df.columns
```

```python
df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
 df['Age'].fillna(df['Age'].mean(),inplace=True)
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
l_sex=LabelEncoder()  #female and male
l_emb=LabelEncoder()  # C,Q,S
df['Sex']=l_sex.fit_transform(df['Sex'])
df['Embarked']=l_emb.fit_transform(df['Embarked'])
```

```python
 df.head()
```

```python
x=df.drop(['Survived'],axis=1)
y=df['Survived']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
import pickle 
pickle.dump(model,open('model.pkl','wb'))
pickle.dump(l_sex,open('l_sex.pkl','wb'))
pickle.dump(l_emb,open('l_emb.pkl','wb'))
```

```python
model=pickle.load(open('model.pkl','rb'))
l_sex=pickle.load(open('l_sex.pkl','rb'))
l_emb=pickle.load(open('l_emb.pkl','rb'))
```
